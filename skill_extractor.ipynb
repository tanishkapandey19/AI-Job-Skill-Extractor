```python
%pip install pandas
import pandas as pd

# 1. A sample Job Description for an internship
job_text = """
We are looking for a Data Science Intern. The candidate must be skilled in Python 
and have a basic understanding of Machine Learning. Experience with GitHub and 
SQL is preferred. Knowledge of Tableau for data visualization is a plus.
"""

# 2. Our 'Skill Bank'—what the AI is searching for
skill_bank = ["Python", "Machine Learning", "SQL", "GitHub", "PowerBI", "Tableau", "Java", "C++"]

# 3. Logic: Find which skills from the bank are in the job text
found_skills = [skill for skill in skill_bank if skill.lower() in job_text.lower()]

# 4. Show the results in a table
print("--- Skills Extracted from Job Post ---")
df_results = pd.DataFrame({'Detected Skills': found_skills})
df_results
```
